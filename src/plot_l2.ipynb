```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = '../output/experiment/res/random_complex_flex.csv'
data = pd.read_csv(file)
data = data[data['result'] == "Success"]
```

```python
plt.figure(figsize=(10, 6))

for domain in data['domain'].unique():
    domain_data = data[data['domain'] == domain]
    plt.scatter(domain_data['total_length'], domain_data['executability'], label=domain)

plt.title('Executability by Total Length for Different Domains')
plt.xlabel('Total Length')
plt.ylabel('Executability')

# Step 5: Add a legend to differentiate the lines for each domain
plt.legend(title='Domain')

# Step 6: Show the plot
plt.grid(True)  # Optional: to add grid lines for better readability
plt.show()
```

```python
# Step 2: Create a faceted plot, where each subplot corresponds to one domain
g = sns.FacetGrid(data, col='domain', col_wrap=4, height=3)
g = g.map(plt.scatter, 'total_length', 'executability', marker='o')

# Step 3: Add titles and adjust the layout
g.fig.suptitle('Executability by Plan Length for Different Domains', y=1.02)
g.set_axis_labels('Toal Length', 'Executability')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='total_length', y='executability', scatter_kws={'s': 10}, line_kws={"color": "red"})

# Titles and labels
plt.title('Executability by Total Length with Regression Line')
plt.xlabel('Total Length')
plt.ylabel('Executability')

plt.grid(True)
plt.show()
```
